## Installation

>Create a virtual environment
>>conda create --prefix env python=3.7 -y

> Run the command in terminal
>>pip install mlflow

## Running Mlflow
>User can log parameters, metrics, models, etc.. in mlflow in the codebase.

> To log model, based on the model selected(tensofrlow, sklearn,pytorch), the code for logging will be different.

> User can register the model if needed.

>On running the python file, all these parameters will get stored and will be present in a folder called mlruns.

> Details like model pikle file, requirements.txt etc.. will be logged.

>Once the file execution is completed mlruns folder will be visible in the root directory.

>To visualize the experiment, run the following command
>>mlflow ui


> User can select experiments and click on __Compare__ button to compare them.

> User can use sql like query to select required models from a pool of models.
>>![MlFlow UI](images/mlflow_ui.PNG)

Use the following link https://github.com/devbk007/mlflow-introduction


## Packaging a project in MlFlow way
> Create conda environment and activate it. Then run the following command
>> conda env export > conda.yaml
> Edit conda.yaml file created.
>> Change the name.
>> Remove unwanted dependencies.
>> Make python as dependency prior to pip.

> Create MLproject file.
>> touch MLproject

> Mention name, conda environment and entry points.
>> main has to be an entry point, otherwise an error will be thrown.

### Running the program

> Run the following command.
>> mlflow run .
>>> It failed for me, with error "conda is not recognized as an internal or external command". On running in Anaconda prompt, got succedded.

> Running without creating a fresh conda environment.
>> mlflow run . --no-conda
>>> This command didn't give any error.

> To change parameters and perform experimentation, do in the following way:
>> mlflow run . -P alpha=0.7 -P l1_ratio=0.3 --no-conda
